```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_probability as tfp

T = 50
alpha = 0.2
pr = 0.8

r = tfp.distributions.Bernoulli(probs=pr, dtype=tf.float32).sample(T)
V = np.ndarray((T+1, ))
V[0] = 0.
for t in range(1, T):
    V[t+1] = V[t] + alpha * (r[t] - V[t]) 
```

```python
plt.plot(range(T+1), V)
plt.show()
```
